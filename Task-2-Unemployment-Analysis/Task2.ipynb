This project focuses on analyzing the unemployment rate in India, particularly observing the impact of Covid-19 on employment trends.

## Problem Statement
Unemployment is measured as the percentage of unemployed people in the total labor force.
During the Covid-19 pandemic, India experienced a sharp rise in unemployment rates.

The objective of this project is to analyze unemployment data and understand trends across time.

## Importing Required Libraries
The following libraries are used for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset
The dataset contains unemployment statistics in India, including unemployment rate,
employment rate, and labor participation rate.

```python
df = pd.read_csv("Unemployment in India.csv")
df.head()
```

## Dataset Information
Understanding the structure, columns, and data types of the dataset.

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning

We remove extra spaces from column names and handle missing values if any.

```python
df.columns = df.columns.str.strip()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

## Date Conversion

Converting the date column to datetime format for time-based analysis.

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Unemployment Rate Analysis

Analyzing how the unemployment rate changed over time.

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate Over Time in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

## Impact of Covid-19 on Unemployment

Visualizing unemployment trends during the Covid-19 period.

```python
covid_data = df[df['Date'] >= '2020-03-01']

plt.figure(figsize=(10,5))
sns.lineplot(
    x=covid_data['Date'],
    y=covid_data['Estimated Unemployment Rate (%)']
)
plt.title("Unemployment Rate During Covid-19")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

## Region-wise Unemployment Analysis

Comparing unemployment rates across different regions.

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df
)
plt.xticks(rotation=45)
plt.title("Region-wise Unemployment Rate")
plt.show()
```

## Conclusion

The analysis shows a significant rise in unemployment rates during the Covid-19 pandemic.
The unemployment trend highlights the economic impact of lockdowns and reduced economic activity.

This project demonstrates how data analysis and visualization can help understand real-world economic issues.
