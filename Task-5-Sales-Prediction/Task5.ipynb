## Problem Statement
Sales prediction means estimating future sales based on advertising expenses.
The goal of this project is to predict product sales using factors such as
TV, Radio, and Newspaper advertising.

## Importing Required Libraries
We import the necessary Python libraries for data analysis and machine learning.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the dataset

```python
df=pd.read_csv("Advertising.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning
We remove unnecessary columns and prepare the dataset.

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

## Data Visualization
We visualize the relationship between advertising and sales.

```python
sns.pairplot(df)
plt.show()
```

## Feature and Target Selection
Advertising budgets are features and sales is the target variable.

```python
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
```

## Train-Test Split
The dataset is split into training and testing sets.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Model Training
We use Linear Regression to train the sales prediction model.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

## Model Evaluation
The model performance is evaluated using regression metrics.

```python
from sklearn.metrics import r2_score , mean_absolute_error
y_pred =  model.predict(x_test)
print("R2 Score:",r2_score(y_test,y_pred))
print("Mean Absolute Error:",mean_absolute_error(y_test,y_pred))
```

## Sales Prediction Example
We predict sales based on sample advertising values.

```python
sample_data = [[200, 50, 30]]  # TV, Radio, Newspaper
predicted_sales = model.predict(sample_data)

print("Predicted Sales:", predicted_sales[0])
```

## Conclusion
The model successfully predicts sales based on advertising budgets.

make data-driven marketing decisions.
